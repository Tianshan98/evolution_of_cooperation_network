```python
import networkx as nx
import matplotlib.pyplot as plt
```

We used Barab√°si-Albert algorithm for generating an undirected scale-free network. However, there is another option using scale_free_graph to generate a scale-free directed graph.

```python
n=100 #Number of nodes
m=4 #Number of initial links
seed=100 #Optional, seed for random number generator
plt.figure(figsize = (20,20))
G=nx.barabasi_albert_graph(n, m, seed)
nx.draw(G)
plt.show()
```
